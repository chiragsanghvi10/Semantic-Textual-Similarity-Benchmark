```python
import pandas as pd
import os
import numpy as np
import requests
import json
from sklearn.preprocessing import MinMaxScaler
from scipy import stats
fileDir = os.path.dirname(os.path.realpath('__file__'))
```

```python
data1 = open('/home/chirag/Downloads/Stsbenchmark/stsbenchmark/sts-dev.csv')
d1 = data1.readlines()
data2 = open('/home/chirag/Downloads/Stsbenchmark/stsbenchmark/sts-test.csv')
d2 = data2.readlines()
data3 = open('/home/chirag/Downloads/Stsbenchmark/stsbenchmark/sts-train.csv')
d3 = data3.readlines()
data = d1
print(len(data))
```

```python
df = []
for i in range(len(data)):
    df.append(data[i].split('\t'))
score = []
sentence1 = []
sentence2 = []
for j in range(len(df)):
    score.append(df[j][4])
    sentence1.append(df[j][5])
    b = df[j][6]
    sentence2.append(b.rstrip("\n"))
```

```python
dict = list(zip(sentence1, sentence2, score))
df = pd.DataFrame(dict, columns = ['sentence1', 'sentence2', 'score'])
```

```python
df.head()
```

```python
scaler = MinMaxScaler()
df['score'] = scaler.fit_transform(df['score'].values.reshape(-1,1))
```

```python
r = []
for i in range(len(df)):
    sent1 = df.sentence1[i]
    sent2 = df.sentence2[i]
    tokens = roberta.encode(sent1, sent2)
    features = roberta.extract_features(tokens)
    predictions = roberta.model.classification_heads['sentence_classification_head'](features)
    r.append(float(predictions.item()))
df["roberta"] = r
```

```python
df.describe()
```

```python
df['roberta'] = scaler.fit_transform(df['roberta'].values.reshape(-1,1))
```
